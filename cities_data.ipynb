```python
import pandas as pd
from IPython.display import HTML
```

```python
# Store filepath in a variable
filepath = "Resources/city_weather.csv"
```

```python
# Read CSV
cities_data = pd.read_csv(filepath)
cities_data
```

```python
cities_data = cities_data.rename(columns={"Unnamed: 0":"City ID"})
cities_data
```

```python
#Display dataframe in html table
HTML(cities_data.to_html(classes='table table-striped'))
```

```python
#table to file
html = cities_data.to_html(index=False, classes='table table-striped')
#write html file
text_file = open("data.html", "w") 
text_file.write(html) 
text_file.close() 
```
